```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Classifier comparison
```

```python
classifiers = [
    'kNN (k=10)',
    'Linear SVC',
    'Polynomial SVC',
    'Decision Tree',
    'Random Forest',
    'ADA Boost',
    'Gaussian\nNaive Bayes',
    'Multinomial\nNaive Bayes',
    'Benoulli\nNaive Bayes',
    'Linear\nDiscriminant\nAnalysis',
    'Quadratic\nDiscriminan\nAnalysis'
]

F1 = [0.177, 0.1, 0.222, 0.166, 0.241, 0.205, 0.177, 0.175, 0.067, 0.196, 0.056]
training_time = [0.5, 220, 124, 3, 7, 42, 0.12, 0.05, 0.1, 1, 2]
test_time = [31, 0.012, 21, 0.006, 0.06, 0.144, 0.340, 0.015, 0.034, 0.014, 0.840]
```

```python
results = []
for c, i, j, k in zip(classifiers, F1, training_time, test_time):
    results.append(tuple((i, c, j, k)))
```

```python
results = sorted(results)
```

```python
import seaborn as sns
```

```python
fix, ax1 = plt.subplots(figsize=(16, 8))
ax2 = ax1.twinx()

ax1.plot([x[0] for x in results], label='F1 Score', color='red')
ax2.bar([i for i in range(len(training_time))], [x[2] for x in results], width=0.35, label='Training Time')
ax1.set_ylabel('F1 Score', size=16)
ax2.set_ylabel('Training Time', size=16)
ax1.set_xticklabels([x[1] for x in results], rotation=45, size=16)
ax1.set_xticks([i for i in range(len(training_time))])
ax1.grid(b=False)
ax1.yaxis.grid(True)
ax2.grid(b=False)
ax1.set_xlim(0,10.5)
ax2.set_xlim(0,10.5)
ax1.set_ylim(0, 0.3)
ax2.set_ylim(0, 250)
ax1.legend(loc=2, fontsize=14)
ax2.legend(loc=0, fontsize=14)

#Label training time
for l, x, y in zip([x[2] for x in results], [i for i in range(len(training_time))], [x[2] for x in results]):
    ax2.text(x + 0.35/2, y*1.01, "{}s".format(l), ha='center', va='bottom')

plt.tight_layout()
    
plt.savefig('../reports/images/charts/classifiers1.pdf', format='pdf', dpi=1200)
```
