```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pylab import *
```

```python
x = np.linspace(0,10,50)
y = x ** 2
```

```python
#Basic figure
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height (range 0 to 1)

ax.plot(x, y, 'r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Basic Plot');
```

```python
#Plot inside plot

fig = plt.figure()

ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.5,0.3,0.3])

ax1.plot(x,y,'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(y,x,'g')
ax2.set_xlabel('y')
ax2.set_ylabel('x');
```

```python
#Subplots 1 plot
fig, ax = plt.subplots()

ax.plot(x,y,'r')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Single Plot');
```

```python
#Subplots 2 plots
fig, axes = plt.subplots(nrows=1,ncols=2)
for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Plot')
    
fig.tight_layout() #Ensure no overlap between plots
```

```python
#Twin axes
fig, ax1 = plt.subplots()

#Plot first on normal left axis
ax1.plot(x, x**2, lw=2, color='b')
ax1.set_ylabel(r"area $(m^2)$", fontsize=14, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")

#Use same x axis
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color='g')
ax2.set_ylabel(r"volume $(m^3)$", fontsize=14, color='g')
for label in ax2.get_yticklabels():
    label.set_color("green")
```

```python
#Simple Contour

alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T

fig, ax = plt.subplots()
ax.contour(Z,  cmap=matplotlib.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1]);
```

```python
#Import 3D class for 3D plots
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
#Surface plots
fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1,2,1, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0, color='red', alpha=0.7)
ax.set_title("Basic 3D")

ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0, antialiased=False)
ax.set_title("surface_plot with color grading and color bar")
cb = fig.colorbar(p, shrink=0.5);
```

```python
#Wire frame plot
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1,1,1, projection='3d')
p = ax.plot_wireframe(X, Y, Z, rstride=4, cstride=4)
ax.set_title("Wire Frame");
```

```python
#Contour plots with projections
fig = plt.figure(figsize=(10,8))

ax = fig.add_subplot(1,1,1, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
cset = ax.contour(X, Y, Z, zdir='z', offset=-np.pi, cmap=matplotlib.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=-np.pi, cmap=matplotlib.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=3*np.pi, cmap=matplotlib.cm.coolwarm)

ax.set_xlim3d(-np.pi, 2*np.pi);
ax.set_ylim3d(0, 3*np.pi);
ax.set_zlim3d(-np.pi, 2*np.pi);
```

```python
#Different viewing angles of 3D plots

fig = plt.figure(figsize=(12,6))

ax = fig.add_subplot(1,2,1, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
ax.view_init(30, 45)

ax = fig.add_subplot(1,2,2, projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
ax.view_init(70, 30)

fig.tight_layout()
```
