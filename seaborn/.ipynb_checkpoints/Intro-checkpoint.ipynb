```python
#Playing around with Seaborn
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_palette("deep", desat = 0.6)
sns.set_context(rc={"figure.figsize": (8,4)})
```

###LOAD DATA

```python
titanic = pd.read_csv('pdacookbook/data/titanic.csv')
```

```python
server = pd.read_csv('pdacookbook/data/serverdata.csv')
```

```python
mlb = pd.read_csv('pdacookbook/data/mlbsalaries.csv')
```

```python
titanic.head()
```

```python
server.head()
```

```python
mlb.head()
```

```python
###HISTOGRAMS
```

```python
plt.hist(titanic.Age.dropna(), bins=25);
```

```python
plt.hist(server.Time, 25, color = sns.desaturate("indianred", 1));
```

```python
plt.hist(mlb.Salary.dropna(), 25);
```

```python
###BOXPLOTS
```

```python
#Specify x=Val, y=Val to choose axis
sns.boxplot(y=titanic.Age.dropna());
```

```python
#Can groupby in boxplot
sns.boxplot(titanic.Age.dropna(), titanic.Sex)
```

```python
sns.boxplot(server.Time);
```

```python
sns.boxplot(y=mlb.Salary);
```

```python
###Kernal Density Estimate Plot
```

```python
sns.kdeplot(titanic.Age.dropna(), shade = True);
```

```python
sns.kdeplot(server.Time, shade = True);
```

```python
sns.kdeplot(mlb.Salary, shade = True);
```

```python
#Distplot combines Histogram + Kdeplot
sns.distplot(titanic.Age.dropna());
```

```python
sns.distplot(server.Time.dropna());
```

```python
sns.distplot(mlb.Salary.dropna());
```

```python
###Violin Plot
```

```python
sns.violinplot(titanic.Age.dropna());
```

```python
sns.violinplot(titanic.Age.dropna(), titanic.Sex);
```

```python
sns.violinplot(mlb.Salary.dropna());
```

```python
###Cumulative Distribution Function
```

```python
sns.kdeplot(titanic.Age.dropna(), cumulative=True);
```

```python
sns.kdeplot(server.Time.dropna(), cumulative=True);
```

```python
sns.kdeplot(mlb.Salary.dropna(), cumulative=True);
```
